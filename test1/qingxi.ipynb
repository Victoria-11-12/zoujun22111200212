```python
# 1. 删除object列有空值的行
movie2_clean = movie2.dropna(subset=movie2.select_dtypes(include='object').columns, how='any')

# 2. 重置索引（drop=True表示丢弃旧索引）
movie2_clean = movie2_clean.reset_index(drop=True)

# 3. 检查是否还有空值
print(movie2_clean.isnull().sum())
```

```python

movie2_clean.isnull()
```

```python
# 1. 选择所有数值列（int和float类型）
movie2_median = movie2_clean.select_dtypes(include=['int', 'float'])

# 2. 计算每列的中位数
column_medians = movie2_median.median()

# 3. 用中位数填充每列的空值
filled_movie2_median = movie2_median.fillna(column_medians)

# 4. 将填充后的数值列放回原DataFrame
movie2_clean[movie2_median.columns] = filled_movie2_median
```

```python
movie2_clean.to_excel('movie2_clean.xlsx', index=False)
```
